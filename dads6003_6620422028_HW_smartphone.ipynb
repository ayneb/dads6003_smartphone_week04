```python
df2 = df2[ df2["Current_OS"] != "IOS และ Andriod" ]     # ตอนนี้ df2  จะไม่มี "IOS และ Andriod" ละ เหลือแต่ IOS กับ Android
df2
```

```python
df2["Current_OS"] = df2 ["Current_OS" ].replace({"IOS": 1, 'Android': 0})    # ในคอลลัม Current_OS  เหลือแต่ IOS กับ Android ที่ตอนนี้เป็น T F อยู่ มาแทนค่า 0 1
df2
```

```python
df2.info()
```

```python
df_encoded  = df2.replace({True: 1, False: 0})
df_encoded
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Select features (X) and target (y)
X = df_encoded.drop('Current_OS', axis=1)
y = df_encoded['Current_OS']

y
```

```python
# Split the dataset into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)

# Create a logistic regression model
logistic_model = LogisticRegression()

# Train the model
logistic_model.fit(X_train, y_train)

# Test the model
accuracy = logistic_model.score(X_test, y_test)

accuracy
```
